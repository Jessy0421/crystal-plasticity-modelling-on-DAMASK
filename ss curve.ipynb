```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_1 = pd.read_excel('plane_cube.xls')
df_2 = pd.read_excel('plane_cube_EBSD.xls')
df_3 = pd.read_excel('plane_sheet.xls')
df_4 = pd.read_excel('plane_sheet_EBSD.xls')
```

```python
import seaborn as sns


sns.set(style="darkgrid")

plt.figure(figsize=(6, 6))


sns.lineplot(data=df_1, x='true_strain', y='true_stress', color='blue', label='Random cube')
sns.lineplot(data=df_2, x='true_strain', y='true_stress', color='yellow', label='EBSD cube')
sns.lineplot(data=df_3, x='true_strain', y='true_stress', color='green', label='Random sheet')
sns.lineplot(data=df_2, x='true_strain', y='true_stress', color='red', label='EBSD sheet')

plt.title('strain-stress curve')
plt.legend()

plt.xlabel('true strain')
plt.ylabel('true stress(Pa)')

plt.savefig('plane_com.png')

plt.show()
```
